## Aglomerative Single Linkage clustering

Agglomerative clustering is a suite of algorithms all based on the same idea. The fundamental idea is that you start with each point in it's own cluster and then, for each cluster, use some criterion to choose another cluster to merge with. Do this repeatedly until you have only one cluster and you get get a hierarchy, or binary tree, of clusters branching down to the last layer which has a leaf for each point in the dataset. The most basic version of this, single linkage, chooses the closest cluster to merge, and hence the tree can be ranked by distance as to when clusters merged/split.

**Code your own Aglomerative Single Linkage clustering algorithm**!:

- Fill in the class 
- During the process, keep track of the cluster merges by saving a (num_samples-1,4) np.array being a linkage matrix in scypy format (to use their function to plot dendrogram: `scipy.cluster.hierarchy.dendrogram`). check documentation.

```python
from scipy.spatial import distance
import scipy
from copy import copy
```

```python
class my_SingleLinkageAglomerativeClustering():
    
    def __init__(self, metric="euclidean", n_clusters=3):
        '''
        Attributes:
        
        metric_: {"euclidean","precomputed"}
            the distance to be used
            if precomputed then X is distance matrix
        n_clusters: integer
            number of clusters to return 
        linkage_matrix_: (n-1, 4) np.array
            in the same format as linkage  
        labels_: integer np.array
            label assigment
        hierarchy_: list of np.array
            each array corresponds to label assigment
            at each level (number of clusters)
            hierarchy_[0]=np.array(list(range(n)))
        '''
        self.metric_ = metric
        self.n_clusters_ = n_clusters
        self.linkage_matrix_ = None
        self.labels_ = None # list of the names of the clusters
        self.hierarchy_ = None
    
    def find_min(self, dist_matrix):
        """Find the row and the col of the min>0 of dist_matrix
        Arg:
        
        dist_matrix: a squared matrix with the distances to each point
        [0      dist01 dist02
         dist01   0    dist 12
         dist02 dist12    0]
         
         Returns:
         
         dist_min: min>0 of dist_matrix
         row_min, col_min: the indexes of dist_min"""
        row_min = 0
        col_min = 1 #can't have equality
        
        dist_min = dist_matrix.max()
        for i in range(dist_matrix.shape[0]):
            for j in range(dist_matrix.shape[1]):
                if (0 < dist_matrix[i][j] < dist_min):
                    dist_min = dist_matrix[i,j]
                    row_min = i
                    col_min = j
                    # we need to change the key and row lines, and change hierarchy to dic_cluster[row] 
        #i_min and j_min are keys !!!
        
        return dist_min, row_min, col_min
    
    def update_dic(self, dic_clusters, row_min, col_min, n, nb_iter):
        """Update dic_clusters by removing the unwantend keys and adding the new cluster"""
        
        count_cluster = dic_clusters[row_min][1] + dic_clusters[col_min][1]
        old_keys = list(dic_clusters.keys())
        
        new_dict = {}
        
        old_keys.remove(row_min)
        old_keys.remove(col_min)
        
        key_min = min(row_min, col_min)
        key_max = max(row_min, col_min)
        
        # removing the old keys
        for key in old_keys:    
            if key<key_min:
                new_dict[key] = dic_clusters[key]
            elif key_min<key<key_max:
                new_dict[key-1] = dic_clusters[key]
                
            else:
                new_dict[key-2] = dic_clusters[key]
                
        # we add the new cluster
        new_dict[n-nb_iter-2] = [n+nb_iter, count_cluster] # indexes start at 0
        
        
        return new_dict

    def update_matrix(self, dist_matrix, index_min, index_max, nb_iter, X, metric, n, dic_clusters):
        """Update the distance matrix
        Remove two rows and cols and add a new one"""
        dist_matrix = np.delete(dist_matrix, index_max, axis=0) # row
        dist_matrix = np.delete(dist_matrix, index_min, axis=0)
        dist_matrix = np.delete(dist_matrix, index_max, axis=1) # col
        dist_matrix = np.delete(dist_matrix, index_min, axis=1)

        # augmtent dist_matrix to add cluster to other cluster distance
        new_dist = np.zeros((dist_matrix.shape[0]+1, dist_matrix.shape[0]+1))

        new_dist[:-1,:-1] = dist_matrix
            
        row = np.zeros(n-nb_iter-1)
        cluster1 = np.where(self.hierarchy_[nb_iter+1] == (n+nb_iter))
        keys = list(dic_clusters.keys())
        
        # compute the new row
        for i in keys:
            
            cluster2 = np.where(self.hierarchy_[nb_iter+1] == dic_clusters[i][0])
            row[i] = self.compute_dist(cluster1, cluster2, X, metric)
        new_dist[-1,:] = row
        new_dist[:,-1] = row
        new_dist[-1,-1] = 0
            
        return new_dist
    
    def fit(self, X):
        """ Create a hierarchy of clusters
        
        Parameters:
        -----------
        X: (n, p) np.array
            Data matrix
        
        Returns:
        -----
        self: my_SingleLinkageAglomerativeClustering
            to have access to labels_
        """
        # if it's not precomputed compute the distance matrix
        # using from scipy.spatial import distance 
        
        # HINT:
        # You can use a minimum spanning tree and add merge in increasing order
        # or modifying the distance matrix 
        # (add row/column for new clusters and remove/put zero in old row/colums)
        
        # keep track of merges in linkage_matrix_ and labels in hierarchy_
        
        # update labels_ from the hierarchy level selected by n_clusters_ 
        n = X.shape[0]
        self.hierarchy_ = [np.array(list(range(n)))] # [ [0, ... , n-1] ]
        #self.labels = np.array(2*n-1)
        self.linkage_matrix_ = np.zeros((n-1, 4))
         
        dist_matrix = distance.cdist(X, X, metric=self.metric_)
        
        # This dict will associate to each line in
        # dist matrix its corresponding clusters and the number of points associated
        # At first, the matrix is (n,n) so each cluster i is on row i with 
        # only 1 point
        # {mat_row : [cluster, count_cluster], ... }
        dic_clusters = {i:[i, 1] for i in range(n)} 
        
        nb_iter = 0
        max_iter = n-1
        while nb_iter < max_iter:
             
            print(f'Iteration {nb_iter} on {max_iter} ...')    
            # look for minimum between the distances
            # i_min and j_min are keys !!!
            dist_min, i_min, j_min = self.find_min(dist_matrix)
            # count the number of points in the cluster
            count_cluster = dic_clusters[i_min][1] + dic_clusters[j_min][1]
            cluster1, cluster2 = dic_clusters[i_min][0], dic_clusters[j_min][0]
            index_min = min(i_min, j_min)
            index_max = max(i_min, j_min)
            
            # update linkage_matrix
            self.linkage_matrix_[nb_iter] = [cluster1, cluster2, copy(dist_min), count_cluster]
            
            # update hierarchy_: get old state and update the new cluster
            snapshot = copy(self.hierarchy_[nb_iter])
            indexes1 = np.where(snapshot == cluster1)
            indexes2 = np.where(snapshot == cluster2)
            indexes = np.concatenate((indexes1, indexes2), axis=1)
            snapshot[indexes] = nb_iter + n
            self.hierarchy_.append(snapshot)
            
            # update dict_clusters, we erase cluster1 and cluster2, new cluster is 
            # added at the end of the matrix
                       
            dic_clusters = self.update_dic(dic_clusters, i_min, j_min, n, nb_iter)
            
            # update distances
            dist_matrix = self.update_matrix(dist_matrix, index_min, index_max, nb_iter,
                                             X, self.metric_, n, dic_clusters)
            nb_iter += 1
            

            
    def compute_dist(self, cluster1, cluster2, X, metric):
        """Find the distance between two clusters
        It s the minimum of the distance between each point in each cluster"""
        l_dist = []
        for pt1 in X[cluster1]:
            for pt2 in X[cluster2]:
                l_dist.append(distance.cdist([pt1], [pt2]))
        dist = min(l_dist)
        
        return dist
    
    def plot_dendrogram(self):
        '''Use self.linkage_matrix_ in `scipy.cluster.hierarchy.dendrogram` 
        to plot the dendrogram of the hierarchical structure
        '''     
        scipy.cluster.hierarchy.dendrogram(self.linkage_matrix_)
```

## DBSCAN

DBSCAN is a density based algorithm -- it assumes clusters for dense regions. It is also the first actual clustering algorithm we've looked at: it doesn't require that every point be assigned to a cluster and hence doesn't partition the data, but instead extracts the 'dense' clusters and leaves sparse background classified as 'noise'. In practice DBSCAN is related to agglomerative clustering. As a first step DBSCAN transforms the space according to the density of the data: points in dense regions are left alone, while points in sparse regions are moved further away. Applying single linkage clustering to the transformed space results in a dendrogram, which we cut according to a distance parameter (called epsilon or `eps` in many implementations) to get clusters. Importantly any singleton clusters at that cut level are deemed to be 'noise' and left unclustered. 

Apply it to the test_data, how do you tune the parameters?

```python
from sklearn.cluster import DBSCAN

clustering_DBSCAN_0 = DBSCAN(eps=0.4, min_samples=2).fit_predict(test_data_0)
clustering_DBSCAN_1 = DBSCAN(eps=0.03, min_samples=20).fit_predict(test_data_1)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].scatter(test_data_0.T[0], test_data_0.T[1], c=clustering_DBSCAN_0 , **plot_kwds)
ax[0].set_title('Toy Dataset', size=16)

ax[1].scatter(test_data_1.T[0], test_data_1.T[1], c=clustering_DBSCAN_1 , **plot_kwds)
ax[1].set_title('Noised Toy Dataset', size=16)

plt.show();
```

## HDBSCAN

HDBSCAN is a recent algorithm developed by some of the same people who write the original DBSCAN paper. Their goal was to allow varying density clusters. The algorithm starts off much the same as DBSCAN: we transform the space according to density, exactly as DBSCAN does, and perform single linkage clustering on the transformed space. Instead of taking an epsilon value as a cut level for the dendrogram however, a different approach is taken: the dendrogram is condensed by viewing splits that result in a small number of points splitting off as points 'falling out of a cluster'. This results in a smaller tree with fewer clusters that 'lose points'. That tree can then be used to select the most stable or persistent clusters. This process allows the tree to be cut at varying height, picking our varying density clusters based on cluster stability.

```python
import hdbscan

clustering_HDBSCAN_0 = hdbscan.HDBSCAN().fit_predict(test_data_0)
clustering_HDBSCAN_1 = hdbscan.HDBSCAN().fit_predict(test_data_1)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].scatter(test_data_0.T[0], test_data_0.T[1], c=clustering_HDBSCAN_0 , **plot_kwds)
ax[0].set_title('Toy Dataset', size=16)

ax[1].scatter(test_data_1.T[0], test_data_1.T[1], c=clustering_HDBSCAN_1 , **plot_kwds)
ax[1].set_title('Noised Toy Dataset', size=16)

plt.show();
```

#### Parameter tuning : 

- On commence par des contraintes très faibles sur la construction des clusters : eps = 1 et min_sample = 2, 
ie le voisinage d'un point est très large et il suffit de deux points proches pour construire un nouveau
cluster.
Pour ces réglages, on obtient un unique cluster contenant tous les points. Le réglage est trop grossier
par rapport à la densité des points.

- Augmenter min_sample n'a aucun effet car tous les points sont dans le voisinage de tous les autres, et 
plus généralement, on s'attend à ce que si la majorité des points ont au moins p(eps) voisins, alors faire
varier min_sample entre 2 et p(eps) ne changera pas sensiblement le résultat

- On diminue progressivement eps pour voir apparaître différents clusters. A eps = 0.8, on voit apparaître 
trois clusters, min_sample n'a aucun impact, l'augmenter jusqu'à 15 ne conduit qu'à casser un cluster en
labelisant certains de ses points en outliers.

- On diminue encore eps jusqu'à 0.5, cela n'affecte pas le clustering (toujours 3 clusters) mais seulement
le seuil de min_sample à partir duquel certains clusters sont cassés, min_sample = 7.

- On diminue ensuite eps à 0.4, et on peut alors obtenir deux clustering différents selon la valeur de 
min_cluster : pour min_cluster = 2, on obtient la même séparation que précédement mais pour min_cluster 
= 5, on obtient 4 différents clusters. Par contre le choix est très sensible, pour min_cluster = 6, 
on casse deux clusters et on obtient un multitude d'outliers. Ensuite, augmenter min_cluster augmente
le nombre d'outliers

- Pour eps = 0.3 et min_cluster = 2, on obtient 4 clusters, avec 2 outliers, qui appartenaient à des 
clusters plus isolés mais moins denses. Les clusters sont déformés à partir de min_sample = 5.

- Enfin, à partir de eps = 0.2, il y a une multitude de sous clusters qui sont formés, et beaucoup 
d'outliers, le réglage est trop fin par rapport à la densité des points.

- En conclusion on commence par un eps trop grand et min_sample = 2, puis on diminue progressivement eps 
tout en essayant plusieurs valeurs de min_sample jusqu'à obtenir un résultat satisfaisant. On conserve
eps = 0.4 et min_cluster = 2

- Pour le deuxième jeu de données plus dense, on opère de la même manière, on commence avec eps = 1 et 
min_sample = 2, ce qui réunit tous les points dans un même cluster. Il faut attendre eps = 0.1 pour 
observer deux points labellisés différement, puis eps = 0.03 pour voir apparaître des clusters. 
Diminuer encore eps n'améliore pas les résultats. Cela commence par scinder les différents clusters formés
puis finit par tout labelliser en outliers.

- Il y a beaucoup de points qui correspondent à du bruit, et si min_sample est trop faible, le bruit 
construit un pont entre les différents clusters et empêche l'algorithme d'être efficace.

- On fixe eps = 0.03 et on essaye différents min_sample. Pour 20, le résultat est satisfaisant, 
plusieurs clusters sont identifiés et le bruit est correctement labellis2. Néanmoins le choix de eps est
très sensible, car pour eps = 0.01, on obtient presque que du bruit.

### Color compression
One interesting application of clustering is in color compression within images. 
For example, imagine you have an image with millions of colors.
In most images, a large number of the colors will be unused, and many of the pixels in the image will have similar or even identical colors.

Get a simplified 10-colored version of the following image by applying k-means. Plot both images.

```python
from sklearn.datasets import load_sample_image
china = load_sample_image("china.jpg")

fig = plt.figure(figsize=(8, 6))
plt.imshow(china);

```

```python
h,w = china.shape[:2]
array = china.reshape(-1,3)
cluster = KMeans(n_clusters=10)
labels = cluster.fit_predict(array)
china_compressed = cluster.cluster_centers_[labels].reshape(h,w,3)/255

plt.imshow(china_compressed)
```
